Las siguientes librerías nos servirán para realizar el análisis de k-means y así ubicar aquellos puntos que nos serivrán para visualizar en el mapa del territorio nacional los centroides que corresponden a la calidad del agua.

````python
from sklearn.cluster import KMeans
from geopy.geocoders.yandex import Location
from geopy.geocoders import Nominatim 
from geopy.distance import geodesic
````

Realizamos una gráfica de codos para determinar la cantidad de centroides a graficar. 

````python
num_clusters = range(1,50)
scores = list()

for k in num_clusters:
    kmeans = KMeans(n_clusters=k)
    estimator = kmeans.fit(latlong[['LONGITUD']])
    scores.append(estimator.score(latlong[['LONGITUD']]))

plt.figure(figsize=(15,10))
plt.plot(num_clusters, scores)
plt.xlabel('Número de Clusters')
plt.ylabel('Score')
plt.title('Gŕafica de codo')
plt.xticks(rotation=90)
plt.xticks(np.arange(min(num_clusters), max(num_clusters)+1, 1.0))
plt.show()
````

Hemos decidido quedarnos únicamente con 4 centroides, ya que esto nos parece es lo más adecuado considerando la calidad del agua por cada una de las regiones.

````python
kmeans = KMeans(n_clusters = 20, random_state=41)
kmeans.fit(latlong[['LONGITUD','LATITUD']])

centroids = kmeans.cluster_centers_
labels = kmeans.predict(latlong[['LONGITUD','LATITUD']])
latlong["label"] = labels
latlong["semaforo"] = df["SEMAFORO"]

centroids_df = pd.DataFrame(centroids)
centroids_df["latlon"] = list(zip(centroids_df[0], centroids_df[1]))
centroids_df["latlon"] = centroids_df["latlon"].apply(Point)

for label in np.unique(labels):
    centroids_df.loc[label, "color"] = latlong[latlong["label"] == label]["semaforo"].mode()[0]

print("=============Labels================")
print(np.unique(labels))
print("=============Centroides============")
print(centroids_df)

latlong.head()
````

#**3. Importancia de las variables**#
##*Analiza la importancia de las variables a través de la función feature_importances.*##

````python
def plot_feature_importance(importance,names,model_type):

    feature_importance = np.array(importance)
    feature_names = np.array(names)
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    plt.figure(figsize=(10,8))
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    plt.title(model_type + ' FEATURE IMPORTANCE')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
````

````python
plot_feature_importance(rf.feature_importances_,x_train.columns,'RANDOM FOREST')
````
